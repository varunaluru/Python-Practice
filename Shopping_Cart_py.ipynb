```python
#SHOPPING CART

# Get user's name and price
name = input("Enter your name: ")
price = float(input("Enter the price: "))

# Validate price input
if price <= 0:
    print("Invalid price")
else:
    # Store initial price before any changes
    original_price = price

    # Calculate discount based on initial price
    if original_price > 1000:
        discount = original_price * 0.2  # 20% discount for prices over 1000
    else:
        discount = original_price * 0.1  # 10% discount for prices 1000 or less

    # Check for VIP status and apply additional discount
    vip = input("Are you a VIP? (yes/no): ")
    if vip == "yes":
        discount = discount + (original_price * 0.1)  # Additional 10% discount for VIPs

    # Calculate GST
    gst = original_price * 0.05

    # Adjust price if it's less than 500 (e.g., small order fee)
    if price < 500:
        price = price + 100

    # Give feedback based on the final price
    if price > 10000:
        print("YOU SPEND TOO MUCH")
    elif price < 100:
        print("YOU SPEND SMART")

    # Calculate the final total after all adjustments
    final_total = price - discount + gst

    # Print the receipt details
    print("\n--- RECEIPT ---")
    print("Customer Name: ", name)
    print("Price (adjusted): ", price)
    print("Discount Applied: ", discount)
    print("GST (5%):         ", gst)
    print("Total Amount:     ", final_total)
    print("Thank you for shopping")
    print("Have a nice day")
```
