```python
while True:
    # Prompt user for their name
    name = input("Enter your name: ")

    # Loop to get a valid price input
    while True:
        try:
            price = float(input("Enter the price: "))
            # Validate price: must be greater than 0
            if price <= 0:
                print("Invalid price. the price must be greater than 0.")
            else:
                break # Exit loop if price is valid
        except ValueError:
            # Handle non-numeric input
            print("Invalid price. Please enter numbers only.")

    # Store the original price for calculations
    original_price = price

    # Calculate discount based on original price
    if original_price > 1000:
        discount = original_price * 0.2 # 20% discount for prices over 1000
    else:
        discount = original_price * 0.1 # 10% discount for prices 1000 or less

    # Check if the customer is a VIP and apply additional discount
    vip = input("Are you a VIP? (yes/no): ")
    if vip == "yes":
        discount = discount + (original_price * 0.1) # Add another 10% discount for VIPs

    # Calculate Goods and Services Tax (GST)
    gst = original_price * 0.05 # 5% GST

    # Adjust price based on its value (add 100 if less than 500)
    if price < 500:
        price = price + 100

    # Provide spending feedback based on adjusted price
    if price > 10000:
        print("YOU SPEND TOO MUCH")
    elif price < 100:
        print("YOU SPEND SMART")

    # Calculate the final total amount
    final_total = price - discount + gst

    # Print the detailed receipt
    print("\n--- RECEIPT ---")
    print("Customer Name:    ", name)
    print("Price (adjusted): ", price)
    print("Discount Applied: ", discount)
    print("GST (5%):         ", gst)
    print("Total Amount:     ", final_total)
    print("Thank you for shopping")
    print(f"Have a nice day {name}!\n")

    # Ask if the shopping is over
    over = input("Is your shopping over? (yes/no): ").strip().lower()
    if over == "yes":
        print(f"Goodbye {name}!")
        break # Exit the main loop if shopping is over
    elif over == "no":
        print("\n--- Next Customer ---")
        continue # Continue to the next customer
    else:
        print("Invalid input. Closing program.")
        break # Exit if input is invalid
```
