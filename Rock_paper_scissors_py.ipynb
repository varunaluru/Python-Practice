```python

# Rock paper scissors

import random #Imports the random function


options = ("rock","paper","scissors") #The options that were given

players_name = input("Please enter your name: ") #Asks the user their name

computers_choice = random.choice(options)#Asks computer to take a randomly choose out of the options
while True:
  players_choice = input("Please choose Rock, Paper or Scissors: ").lower() #Asks user what choice they want to do
  if players_choice not in options:

    print("Invalid input. Please enter one of the options") #The message if the user doesn't type one of the options
    continue
  else:
    break

if players_choice == computers_choice:#If both the user and the computer pick the same option

  print(f"You chose {players_choice}")

  print(f"Computer chose {computers_choice}")
  print("It's a tie 🤝")

elif players_choice == "rock" and computers_choice == "scissors":#If the user picks rock and the computer picks scissors you win
  print(f"Computer chose {computers_choice}")

  print("You win 🏆")

elif players_choice == "paper" and computers_choice == "rock":#If the user picks paper and the computer picks scissors you win
  print(f"Computer chose {computers_choice}")

  print("You win 🏆")

elif players_choice == "scissors" and computers_choice == "paper":#If the user picks scissors and the computer picks paper you win
  print(f"Computer chose {computers_choice}")

  print("You win 🏆")

else:
  print(f"You chose {players_choice}") #prints the user's option at the end

  print(f"Computer chose {computers_choice}") #prints the computer's option at the end

  print("You lose ❌") #The message that will pop up if you lost or the other statments were not satisfied.
```
