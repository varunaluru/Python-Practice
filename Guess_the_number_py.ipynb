```python
#project 2
import random
secretNumber = random.randint(1,50)
attempt = 1
x = 0

while x != secretNumber and attempt < 6:
  x = int(input("Guess the random number: "))

  if x > secretNumber:
    print("Too high")
    if attempt == 5:
        print()
        print("Sorry that's not the right number.")
        print("The actual number was", secretNumber)
  elif x < secretNumber:
    print("Too low")
    if attempt == 5:
        print()
        print("Sorry you didn't get it.")
        print("The actual number was", secretNumber)
  else:
    print("You got it!")

  print()

  attempt = attempt + 1


print("Thanks for playing!")
```
