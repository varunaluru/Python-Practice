```python
import random
print("""
  ====================
  🎮 H A N G M A N 🎮
  ====================""")
print(" 🎮 Welcome to the Hangman game")
words = [
    "volcano",
    "australia",
    "football",
    "crocodile",
    "number",
    "google",
    "apple",
    "coconut",
    "roblox",
    "suitcase",
    "dinosaur",
    "aura",
    "macbook",
    "paper",
    "photo",
    "portfolio",
    "kingdom",
    "throne",
    "elephant",
    "mystical",
    "interface",
    "programming"

]
hangman_art = [
    """
    +----+
         |
         |
         |
         |
         |
    =========
    """,
    """
    +----+
         |
     o   |
         |
         |
         |
    =======
    """,
    """
    +----+
         |
     o   |
     |   |
         |
         |
    =======
    """,
    """
    +----+
         |
     o   |
    /|   |
         |
         |
    =======
    """,
    """
    +----+
         |
     o   |
    /|\\  |
         |
         |
    =======
    """,
    """
  +----+
         |
     o   |
    /|\\  |
    /    |
         |
    =======
    """,
    """
    +----+
         |
     o   |
    /|\\  |
    / \\  |
         |
    =======
    """,
]

secret_word = random.choice(words)

hidden_word = []

for letter in secret_word:
    hidden_word.append("_")

lives = 6

for item in hidden_word:
    print(item, end=" ")

print()

while lives > 0:

    guess = input(" 🔤 Enter a letter: ")

    guess_correct = False

    for position, letter in enumerate(secret_word):
        if letter == guess:
            hidden_word[position] = letter
            guess_correct = True

    if not guess_correct:
       lives -= 1
       print(f"  ❌ Wrong guess! You have {lives} ❤️ lives left.")
       print(hangman_art[6 - lives])

    for item in hidden_word:
        print(item, end=" ")

    print()

    if "_" not in hidden_word:
        print(" 🎊 Congratulations, you guessed the word! 🏆 ")
        print(f" The word was 🔎 {secret_word}.")
        break

    if lives == 0:
        print(f" 💀 Game over! The word was 🔎 {secret_word}.")
        break
```
